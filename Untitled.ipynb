```python
from matplotlib import pylab
import pylab as p
import matplotlib as plt
import numpy as np
```

```python
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

mySamples = []
mySigmoid = []

# generate an Array with value ???
# linespace generate an array from start and stop value
# with requested number of elements. Example 10 elements or 100 elements.
# 
x = p.linspace(-10,10,10)
y = p.linspace(-10,10,100)

# prepare the plot, associate the color r(ed) or b(lue) and the label 
p.plot(x, sigmoid(x), 'r', label='linspace(-10,10,10)')
p.plot(y, sigmoid(y), 'b', label='linspace(-10,10,100)')

# Draw the grid line in background.
#p.grid()

# Title & Subtitle
p.title('Sigmoid Function')
#p.suptitle('Sigmoid')

# place the legen boc in bottom right of the graph
p.legend(loc='lower right')

# write the Sigmoid formula
#p.text(4, 0.8, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=15)

#resize the X and Y axes
p.gca().xaxis.set_major_locator(p.MultipleLocator(1))
p.gca().yaxis.set_major_locator(p.MultipleLocator(0.1))
 

# plt.plot(x)
p.xlabel('X Axis')
p.ylabel('Y Axis')

# create the graph
p.show()
```
